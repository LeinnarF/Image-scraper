```python
import os
path = 'downloaded_images'
num_images = len(os.listdir(path))
print(f'Total images downloaded: {num_images}')
```

```python
# Check broken images
from PIL import Image
broken_images = []
for img_name in os.listdir(path):
    img_path = os.path.join(path, img_name)
    try:
        img = Image.open(img_path)
        img.verify()  # Verify that it is, in fact an image
    except (IOError, SyntaxError) as e:
        broken_images.append(img_name)

print(f'Total broken images: {len(broken_images)}')
if broken_images:
    print('Broken images:')
    for img in broken_images:
        print(img)
```

```python
# Delete broken images
for img_name in broken_images:
    img_path = os.path.join(path, img_name)
    os.remove(img_path)
print(f'Deleted {len(broken_images)} broken images.')
```
